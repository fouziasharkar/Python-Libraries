### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
team_names = fifa.groupby('Team')
team_names.size()
````

````python
def z_normalization(group,cols):
  for i in cols:
    s = group[i].std() #std
    m = group[i].mean() #meu or mean
    group[i]= (group[i]-m)/s
  return group


team_names.apply(z_normalization,['Passes', 'Passes Completed', 'Attempted Line Breaks', 'Completed Line Breaks'])

team_names.agg({

            'Passes':['min','max',],
            'Yellow Cards':['min','max','mean'],
            'Attempted Line Breaks':['min','max','std'],
            'Possession (%)':'mean'
})
````

###`Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
ipl.head(1)
````

````python
#why i used sorted in the code

'''
For the first row (a-b):

row['batter'] = 'a'
row['non-striker'] = 'b'
sorted(['a', 'b']) results in ['a', 'b'], so the 'pair' column value would be 'a-b'.
For the second row (b-a):

row['batter'] = 'b'
row['non-striker'] = 'a'
sorted(['b', 'a']) results in ['a', 'b'], so the 'pair' column value would be 'a-b'.
'''
````

````python
#using sorted to consider vkohli-dhoni and dhoni-vkohli as same
ipl['pair'] = ipl.apply(lambda row: '-'.join(sorted([row['batter'], row['non-striker']])), axis=1)

# Group by the pair and sum the runs
pair_runs = ipl.groupby('pair').agg(total_runs=('total_run', 'sum')).reset_index()

# Find the most successful batting pair
pair_runs.sort_values(by='total_runs', ascending=False).head()
````
